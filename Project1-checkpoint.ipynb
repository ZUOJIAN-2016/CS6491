## Markowiz portfolio problem
### Get start from a classical form of Markowiz portfolio problem

And the classical form of Markowiz portfolio problem is:  
$$Minimize\quad -\bar{p}^Tx+\mu *x^T\Sigma x $$  
$$s.t. 1^Tx =1,x\geq 0$$  
while p is the mean of relative price change and $\Sigma$ is the devariation of the assets  
And we first consider a simple problem with 4 assets:  

$$\bar{p} = \begin{bmatrix}0.12 \\0.10 \\0.07 \\0.03 \end{bmatrix} \quad \Sigma = \begin{bmatrix} 0.0064  &  0.0008&  -0.0011& 0\\ 0.0008  &  0.0025&  0&0\\-0.0011 & 0 & 0.0004 &0 \\  0&  0&  0&0\end{bmatrix} $$  

And this problem is a QP problem. In this problem, $\mu$ is the element relevant to the return and the risk of the portfolio, that is, if $\mu = 0$, then we don't take the variance of the optimized portfolio into account. And if $\mu \to \infty$, then we don't take the portfolio's mean return into account.  

```python
##classical form portfolio problem
import cvxpy as cp
import numpy as np
import math
import matplotlib.pyplot as plt
import time
# Generate a random non-trivial quadratic program.
n = 4
p = np.array([[0.12,0.10,0.07,0.03]])
print(p.shape[0])
gamma = np.array([[0.0064,0.0008,-0.0011,0],[0.0008,0.0025,0,0],[-0.0011,0,0.0004,0],[0,0,0,0]])
A = np.ones([1,n])
# Define and solve the CVXPY problem.
x = cp.Variable(n)
miu = np.logspace(0,7,100)
#miu = 1000
constraints = [cp.sum(x) == 1,0 <= x]
j = 0
solve_set_x1 = []
solve_set_x2 = []
solve_set_x3 = []
solve_set_x4 = []
solve_return = []
solve_devariation = []
start_time = time.perf_counter()
for j in range(len(miu)):
    prob = cp.Problem(cp.Minimize(- p @ x + miu[j]*cp.quad_form(x, gamma)),constraints)
    prob.solve()
    solve_set_x1.append(x.value[0])
    solve_set_x2.append(x.value[1])
    solve_set_x3.append(x.value[2])
    solve_set_x4.append(x.value[3])
    #print(x.value)
    #print(p@(x.value))
    #print(math.sqrt(devariation@(x.value)))
    solve_return.append(p@(x.value))
    solve_devariation.append(math.sqrt(((x.value).T)@(gamma)@(x.value)))
end_time = time.perf_counter()
'''
prob = cp.Problem(cp.Minimize(miu*cp.quad_form(x, gamma) - p.T @ x),constraints)
prob.solve()
print(p.shape)
print((p.T)@(x.value))
'''
plt.figure()
plt.subplot(211)
# Plot result as Fig.12 on the textbook
plt.plot(solve_devariation,solve_return)
plt.subplot(212)
plt.plot(solve_devariation,solve_set_x1,'g')
plt.plot(solve_devariation,solve_set_x2,'r')
plt.plot(solve_devariation,solve_set_x3,'b')
plt.plot(solve_devariation,solve_set_x4,'y')
print("Elapsed time:",end_time-start_time)
plt.show()
```

### An attempt to solve the problem with only equations 
As the $\Sigma$ is semi-positive deninite, we can formulate the problem into solve the equations below:


```python
#Another way to solve the Markowiz portfolio problem
import numpy as np
import math
import matplotlib.pyplot as plt

n = 4
p = np.array([[0.12],[0.10],[0.07],[0.03]])

gamma = np.array([[0.0064,0.0008,-0.0011,0],[0.0008,0.0025,0,0],[-0.0011,0,0.0004,0],[0,0,0,0]])
A = np.ones([1，n])
miu = np.logspace(0,7,100)
G = np.dot(p,A.T)-2*miu[20]*gamma
G_ = np.linalg.inv(G)
l = np.dot(A.T,G_)
# I manually fill in a vector v that satisfy v*l = 1
v = np.array([[ 0.00000000e+00,-8.88178420e-16,0.00000000e+00,0.03]])
x = np.dot(G_,v.T)

print(x)
```
